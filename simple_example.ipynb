```python
import numpy as np
import matplotlib.pyplot as plt
from src.SpikeWizard import Fitter
plt.style.use('monospace')
```

```python
x = np.linspace(-10, 10, 1000)

def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

noise = np.random.normal(0, 0.1, x.shape)
y = np.sum([gaussian(x, mu, sigma) for mu, sigma in zip(np.random.uniform(-10,10, 5), np.random.uniform(.3, 2, 5))], axis=0) + noise
```

```python
Fitter = Fitter(y, function = 'Lorentzian')
```

```python
fig, ax = plt.subplots()
ax.set_title('Lorentzian Fit')
Fitter.plot_fit(ax)
ax.xaxis.set_visible(False)
ax2 = ax.inset_axes([0, -.3, 1, 0.25], sharex=ax)
ax2.set_title('Residuals')
ax2.semilogy(y  - Fitter.fit, alpha = 0.5, lw = 0, marker = 'o', markersize = 2)
ax2.xaxis.set_visible(False)
```
